Helper function

```python
import os, requests

def brawlstars_get(endpoint: str, params: dict = None) -> dict:
    token = os.environ.get('BRAWLSTARS_TOKEN')
    if not token:
        raise RuntimeError("BRAWLSTARS_TOKEN not set in environment")
    headers = {
        'Authorization': f"Bearer {token}",
        'Accept': 'application/json',
        'User-Agent': 'ColabBrawlClient/1.0'
    }
    resp = requests.get('https://api.brawlstars.com' + endpoint,
                        headers=headers, params=params)
    resp.raise_for_status()
    return resp.json()
```

```python
player_tag = '%239JVU8RC'

profile = brawlstars_get(f'/v1/players/{player_tag}')
battlelog = brawlstars_get(f'/v1/players/{player_tag}/battlelog')

# Join everything on a single json
player_data = {
    'profile': profile,
    'battlelog': battlelog
}
print(player_data)

# Download json
import json
with open('player_data.json', 'w') as f:
    json.dump(player_data, f)
```
